```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import scale as s
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import confusion_matrix,classification_report
import numpy as np
```

```python
data = pd.read_csv("Absenteeism_at_work.csv")
```

```python
data.drop(['ID'],axis = 1,inplace = True)
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
X = data.drop(['Absenteeism time in hours'],axis=1)
Y = data['Absenteeism time in hours']
X = s(X)
Y = s(Y)
```

# Regression

```python
Xtrain, Xtest, Ytrain, Ytest = train_test_split(X,Y,test_size = 0.2)
```

```python
lr = LinearRegression()
```

```python
lr.fit(Xtrain,Ytrain)
```

```python
ypreds = lr.predict(Xtest)
print(round(mean_squared_error(Ytest,ypreds),2))
```

```python
rfr = RandomForestRegressor(n_estimators = 100,max_depth = 3)
rfr.fit(Xtrain,Ytrain)
print(round(mean_squared_error(Ytest,rfr.predict(Xtest)),2))
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
data['Acceptable Absentee'] = data['Absenteeism time in hours']<15
```

```python
data['Acceptable Absentee'].value_counts()
```

```python
classification_y = data['Acceptable Absentee']
classif_x_train,classif_x_test,classif_y_train,classif_y_test = train_test_split(X,classification_y,test_size = 0.2)
```

```python
svcmodel = SVC(kernel = 'sigmoid')
svcmodel.fit(classif_x_train,classif_y_train)
```

```python
print("Acuracy of SVM model : {}%".format(round(accuracy_score(classif_y_test,svcmodel.predict(classif_x_test))*100,2)))
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg = LogisticRegression()
logreg.fit(classif_x_train,classif_y_train)
print("Acuracy of SVM model : {}%".format(round(accuracy_score(classif_y_test,logreg.predict(classif_x_test))*100,2)))
```
